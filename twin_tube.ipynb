### Creating an ***id*** column for unique row identification

#### Create an ID column in the dataframe 

From the result of the ***nunique() function*** on the dataframe, we could notice that there is ***primary key*** column to identify each record in the dataset. To make our access to the data easy, we will create the the ID column below. We can do this in multiple ways which include:

- Create a new column using ***pd.factorize()*** function
- Reset the ***index column*** to serve as the ID column
- Creating a new 'id' column and assign it the values of index or (index+1) if we intendnot to start from 0

For the purpose of simplicity, we would use the option 3 to get ID column ready for use.

```python
df2['id'] = df2.index+1
```

```python
df2.head()
```

### Shift the column "***id***" to the first position

```python
# pop the column 'id' and assign to variable first_name
first_column = df2.pop('id')
  
# insert column using insert(position,column_name,
# first_column) function
df2.insert(0, 'id', first_column)
```

### Applying melt() function to prepare the data for time series forecasting

```python
melt = df2.melt(id_vars=['id', 'Mode', 'NLC', 'ASC', 'Station', 'Coverage', 'year', 'day', 'dir', 'Total', 'Early', 'AMPeak', 'InterPeak', 'PMPeak', 'Evening', 'Late'], var_name='15mins_interval', value_name='Count')
melt.head()
```

```python
melt['15mins_interval'] = melt['15mins_interval'].str[:4]
melt.head()
```

```python
melt['15mins_interval'] = melt['15mins_interval'].astype(str).str[:2] + ':' + melt['15mins_interval'].astype(str).str[2:]
```

```python
melt.shape
```

```python
melt['15mins_interval'] = pd.to_datetime(melt['15mins_interval'], format='%H:%M').dt.time
```

```python
melt.tail(1000)
```

### Converting the melted data to csv file for easy access

```python
melt.to_csv('melted_data_forecast.csv', index=False)
```
